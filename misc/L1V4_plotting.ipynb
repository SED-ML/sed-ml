## Shaded Area, order, error bars
https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.errorbar.html
https://matplotlib.org/examples/pylab_examples/fill_between_demo.html

````python
x = np.linspace(0, 10, num=10)
xerr = x*0.2*np.random.uniform(size=10)
y = np.sqrt(x)
yerr = y*0.2*np.random.uniform(size=10)

# error data generators
yerr_up = y + yerr
yerr_down = y - yerr
xerr_up = x + 0.2
y_down = y-0.2

y_zero = np.zeros_like(y)
y_sin = np.sin(x)
````

````python
fig3 = plt.figure(figsize=(10, 5))
fig3.suptitle('Figure 3', fontsize=16)
ax1 = plt.subplot2grid(shape=(1, 2), loc=(0, 0))
ax2 = plt.subplot2grid(shape=(1, 2), loc=(0, 1))

ax1.fill_between(x, yerr_down, yerr_up, color="blue", alpha=0.3, label="error_area")
ax1.errorbar(x, y, xerr=xerr, yerr=yerr, color="black", label="errorbar")
ax1.legend()

ax2.fill_between(x, y_zero, y_sin, color="red")
ax2.plot(x, y_sin, color="black", label="area")
ax2.legend()

plt.show()
````

````python
x = np.linspace(0, 2*np.pi, num=10)
y1 = 5 * np.ones_like(x) + 2 * np.random.uniform(size=10)
y2 = y1 + np.sin(x)
````

````python
fig4, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12,4))
fig4.subplots_adjust(hspace=0.3, wspace=0.3)

fig5, (ax4, ax5, ax6) = plt.subplots(1, 3, figsize=(12,4), subplot_kw={'projection':'polar'})
fig5.subplots_adjust(hspace=0.3, wspace=0.3)

ax1.set_title("points")
ax1.plot(x, y1, '-o')
ax1.plot(x, y2, '-o')

# use width and the x position
width=(x[1]-x[0])/3.0
ax2.set_title("bar")
ax2.bar(x-0.5*width, y1, width=width)
ax2.bar(x+0.5*width, y2, width=width)

ax3.set_title("barStacked")
ax3.bar(x, y1, width=2*width)
ax3.bar(x, y2, bottom=y1, width=2*width)

# polar
ax4.set_title("polarPoints")
ax4.plot(x, y1, '-o')
ax4.plot(x, y2, '-o')

ax5.set_title("polarBar")
ax5.set_title("bar")
ax5.bar(x-0.5*width, y1, width=width)
ax5.bar(x+0.5*width, y2, width=width)

ax6.set_title("polarStacked")
ax6.bar(x, y1, width=2*width)
ax6.bar(x, y2, bottom=y1, width=2*width)
plt.show()
````

## polarPoints

````python
import numpy as np
import pylab as plt

fig = plt.figure()
ax = plt.axes(polar=True)

r =  np.array([5.31,5.29,5.25,5.19,5.09,4.92,4.67,4.27,3.75,3.56])
theta = 8*np.pi/360 * np.array(list(range(0, 100, 10)))

ax.plot(theta, r, "ro")
ax.fill_between(theta, r, color="blue", alpha=0.1)
ax.errorbar(theta, r, yerr=1, xerr=.1, capsize=0)

plt.show()
````

## Styles
https://matplotlib.org/api/markers_api.html

````python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8,12))
markers = {
    'None': None,
    'Square': 's',
    'Circle': 'o',
    'Diamond': 'D',
    'XCross': 'x',
    'Plus': '+',
    'Star': '*',
    'TriangleUp': '^',
    'TriangleDown': 'v',
    'TriangleLeft': '<',
    'TriangleRight': '>',
    'HDash': '_',
    'VDash': '|',
}

linestyles = {
    'None': 'None',
    'Solid': '-',
    'Dash': '--',
    'Dot': ':',
    'DashDot': '-.',
}

x = np.linspace(0, 10, num=11)
y = np.sqrt(x) + 10

for k, key in enumerate(markers.keys()):
    marker = markers[key]
    ax1.plot(x, y-k, color="black", marker=marker, label=key, markersize=10)
ax1.legend()
ax1.set_xlim(0,15)

for k, key in enumerate(linestyles.keys()):
    linestyle = linestyles[key]
    ax2.plot(x, y-k, color="black", label=key, marker='s',
             linestyle=linestyle, linewidth=2)
ax2.legend()
ax2.set_xlim(0,15)

for k in range(10):
    ax3.plot(x, y-k, "-o", markersize=k, linewidth=k, label="{}".format(k))
ax3.legend()
ax3.set_xlim(0,15)

'''
marker 	unknown
markeredgecolor 	color
markeredgewidth 	float
markerfacecolor 	color
markerfacecoloralt 	color
markersize 	float
'''
````
