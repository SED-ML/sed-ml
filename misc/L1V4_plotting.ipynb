````python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
````

````python
x = np.linspace(0, 10, num=50)
y1 = np.power(x, 2)
y2 = x
y3 = np.sqrt(x)
y4 = np.sin(x)
````

## Shaded Area, order, error bars
https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.errorbar.html
https://matplotlib.org/examples/pylab_examples/fill_between_demo.html

````python
x = np.linspace(0, 10, num=10)
xerr = x*0.2*np.random.uniform(size=10)
y = np.sqrt(x)
yerr = y*0.2*np.random.uniform(size=10)

# error data generators
yerr_up = y + yerr
yerr_down = y - yerr
xerr_up = x + 0.2
y_down = y-0.2

y_zero = np.zeros_like(y)
y_sin = np.sin(x)
````

```
<listOfOutputs>
    <plot2d id="plot_errorbar" legend="True">
        <listOfCurves>
            <curve id="cy_err" name="errorbar" order="1" xDataReference="x" yDataReference="y" xError="xerr" yError="yerr style="style0"/>
        </listOfCurves>
        <listOfShadedAreas>
            <shadedArea id="cy_err_area" name="error_area" order="2" xDataReference="x" yDataReferenceFrom="yerr_down" yDataReferenceTo="yerr_up" style="style1">
        </listOfShadedAreas>
    </plot2d>
    <plot2d id="plot_area" legend="True">
        <listOfShadedAreas>
            <shadedArea xDataReference="x" yDataReferenceFrom="yerr_down" yDataReferenceTo="yerr_up" style="style2">
        </listOfShadedAreas>
    </plot2d>
</listOfOutputs>
```

````python
fig3 = plt.figure(figsize=(10, 5))
fig3.suptitle('Figure 3', fontsize=16)
ax1 = plt.subplot2grid(shape=(1, 2), loc=(0, 0))
ax2 = plt.subplot2grid(shape=(1, 2), loc=(0, 1))

ax1.fill_between(x, yerr_down, yerr_up, color="blue", alpha=0.3, label="error_area")
ax1.errorbar(x, y, xerr=xerr, yerr=yerr, color="black", label="errorbar")
ax1.legend()

ax2.fill_between(x, y_zero, y_sin, color="red")
ax2.plot(x, y_sin, color="black", label="area")
ax2.legend()

plt.show()
````
